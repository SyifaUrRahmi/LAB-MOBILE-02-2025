```python
# prompt: load and training data with delimeter ";", and then load testing

# Load training data
training_data = pd.read_csv("/content/sisfo-uh-introduction-of-kaggle-competition/Training.csv", delimiter=";")

# Load testing data
testing_data = pd.read_csv("/content/sisfo-uh-introduction-of-kaggle-competition/Testing.csv", delimiter=";")

# Now you can work with training_data and testing_data
# Example: Print the first 5 rows of the training data
print(training_data.head())

# Example: Print some info about the testing data
print(testing_data.head())
```

```python
print(training_data.head())
print(testing_data.head())
print(training_data.info())
print(testing_data.info())
```

```python
X_train = training_data['menit']  # Semua kolom kecuali label
y_train = training_data["watt"]                 # Kolom label

X_test = testing_data['waktu']
```

```python
# prompt: train with linear regression

# Reshape X to a 2D array
X_train = X_train.values.reshape(-1, 1)
y_train = y_train.values.reshape(-1, 1)
X_test = X_test.values.reshape(-1, 1)

from sklearn.linear_model import LinearRegression
# Initialize and train the linear regression model
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
# prompt: make prediction on the test set

# Make predictions on the test set
y_pred = model.predict(X_test)

# Print the predictions
y_pred
```

```python
import pandas as pd

# Pastikan y_pred menjadi array 1D
y_pred_flat = y_pred.flatten()  # atau bisa juga pakai y_pred.ravel()

# Buat DataFrame untuk menyimpan ke CSV
df_pred = pd.DataFrame({"Predicted": y_pred_flat})

# Simpan ke CSV
df_pred.to_csv("tester.csv")
```
